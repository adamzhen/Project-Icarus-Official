## Solar Probe Cup (SPC)

#### Velocity

```python
import numpy as np
from matplotlib import pyplot as plt
import math
import pyspedas
from pytplot import tplot

spc_vars = pyspedas.psp.spc(trange=['2021-04-24', '2021-04-30'], datatype='l3i', level='l3', get_support_data=True, notplot=True, no_update=False)
t1 = spc_vars['psp_spc_vp_fit_RTN']['x']
v1 = spc_vars['psp_spc_vp_fit_RTN']['y']
t2 = spc_vars['psp_spc_vp_moment_RTN']['x']
v2 = spc_vars['psp_spc_vp_moment_RTN']['y']
velocities1 = []
velocities2 = []
for row in v1:
    total = 0
    for dir in row:
        if np.isnan(dir):
            continue
        total += dir**2
    total = math.sqrt(total)
    if total==0:
        total = row[0]
    velocities1.append(total)
for row in v2:
    total = 0
    for dir in row:
        if np.isnan(dir):
            continue
        total += dir**2
    total = math.sqrt(total)
    if total==0:
        total = row[0]
    velocities2.append(total)
fig, ax = plt.subplots(figsize=(12, 4))
ax.scatter(t1, np.array(velocities1) - np.array(velocities2), label="vp_fit", s=1, alpha=0.5)
#ax.scatter(t2, np.array(velocities2), label="vp_moment", s=1, alpha=0.5)
ax.set_ylabel("Velocity of protons (km/s)")
ax.legend()
```

```python
spc_vars = pyspedas.psp.spc(trange=['2021-04-24', '2021-04-30'], datatype='l3i', level='l3', get_support_data=True, notplot=False, no_update=False)
print(spc_vars)
tplot(['psp_spc_vp_moment_RTN_deltalow', 'psp_spc_vp_moment_RTN_deltahigh', 'psp_spc_vp_fit_RTN_uncertainty'])
tplot(['psp_spc_vp_moment_RTN','psp_spc_vp_fit_RTN'])
tplot(['psp_spc_vp_moment_SC','psp_spc_vp_fit_SC'])
```

#### Temperature

```python
tplot(['psp_spc_wp_moment_deltalow', 'psp_spc_wp_moment_deltahigh', 'psp_spc_wp_fit_uncertainty'])
tplot(['psp_spc_wp_moment','psp_spc_wp_fit'])
```

With this thermal speed data, I noticed huge spikes in the wp_fit graph up to 1000 km/s, whereas the wp_moment graph stayed relatively consistent within 20-200 km/s. So I used T = (mv^2)/2k to calculate example temperatures: 150 km/s turned out to be around 1.36 million Kelvin, which is reasonably accurate to the actual temperature of the corona (and the PSP is indeed inside the corona at this time), but 1000 km/s was a whopping 60.57 million Kelvin, which I think is likely some sort of error as that is way hotter than even the sun's core (15 million Kelvin). As a result, I decided that it'd be better to use the wp_moment data.
